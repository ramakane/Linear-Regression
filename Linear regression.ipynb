```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as nm
```

```python
data=pd.read_csv("headbrain.csv")
```

```python
X_train=data[:200]
X_test=data[200:]
```

```python
X=X_train["Head Size(cm^3)"].values
Y=X_train["Brain Weight(grams)"].values
```

```python
X_mean=nm.mean(X)
Y_mean=nm.mean(Y)
```

```python
num=0
dem=0
m=len(X)

for i in range(m):
    num+=(X[i]-X_mean)*(Y[i]-Y_mean)
    dem+=(X[i]-X_mean)**2

b1=num/dem

b0=Y_mean-b1*X_mean

print(b0,b1)
```

```python
x=nm.linspace(min(X),max(X))
y=b0+b1*x
print(x,y)
```

```python
plt.plot(x,y, color='YELLOW', label='Regression Line')
plt.scatter(X, Y, c='GREEN', label='Scatter Plot headsize vs brain wt')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
X1=X_test["Head Size(cm^3)"].values
Y1=X_test["Brain Weight(grams)"].values
X1_mean=nm.mean(X1)
Y1_mean=nm.mean(Y1)
```

```python
m1=len(X1)
sse=0
sst=0
for i in range(m1):
    y_pred=b0+b1*X1[i]
    sse+=(y_pred-Y1[i])**2
    sst+=(Y1[i]-Y1_mean)**2
    scorer2 = 1 - (sse/sst)
mse=sse/m    
rmse=nm.sqrt(mse)
print("Mean sqaure error is",mse)
print("Root Mean sqaure error is",rmse)
print("R^2 score for training data is",scorer2)
```
